POLYNOMIAL LINEAR REGRESSION: To handle non linear data(data that cant be plotted in straight line) and raise the degree to get a curved line
Underfitting: A statistical model or a machine learning algorithm is said to have underfitting when a model is too simple to capture data complexities.
It represents the inability of the model to learn the training data effectively result in poor performance both on the training and testing data.
Overfitting:A statistical model is said to be overfitted when the model does not make accurate predictions on testing data. When a model gets trained with so much data, it starts learning from the noise and inaccurate data entries in our data set

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
dataset=pd.read_csv("https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Position_Salaries.csv")
```

```python
x=dataset.iloc[:,1:2].values
x
```

```python
y=dataset.iloc[:,2].values
y
```

SIMPLE LINEAR REGRESSION

```python
from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression()
lin_reg.fit(x,y) #summa (dummy)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg2=PolynomialFeatures(degree=2)
x_poly2=poly_reg2.fit_transform(x)
lin_reg_2=LinearRegression()
lin_reg_2.fit(x_poly2,y)
```

CUBIC POLYNOMIAL

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg3=PolynomialFeatures(degree=3)
x_poly3=poly_reg3.fit_transform(x)
lin_reg_3=LinearRegression()
lin_reg_3.fit(x_poly3,y)
```

```python
plt.scatter(x,y,color='red')
plt.plot(x,lin_reg.predict(x),color='blue')
plt.title('Linear Regression')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
plt.scatter(x,y,color='red')
plt.plot(x,lin_reg_2.predict(x_poly2),color='blue')
plt.plot(x,lin_reg_3.predict(x_poly3),color='green')
plt.title('Polynomial Linear Regression')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
data=pd.read_csv('https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Real%20estate.csv')
data
```

```python
sns.displot(data=data,x='Y house price of unit area',kde=True,color='Blue')
```

```python
x=data.drop('Y house price of unit area',axis=1)
x
```

```python
y=data['Y house price of unit area']
y
```

```python
polynomial_reg= PolynomialFeatures(degree=2,include_bias=False)
```

```python
poly_feat=polynomial_reg.fit_transform(x)
poly_feat.shape
```

```python
x_train,x_test,y_train,y_test=train_test_split(poly_feat,y,test_size=0.3,random_state=101) #usual randomstate number=42
```

```python
polymodel=LinearRegression()
```

```python
polymodel.fit(x_train,y_train)
```

```python
y_pred=polymodel.predict(x_test)
```

```python
pd.DataFrame({'Y_test':y_test,'Y_pred':y_pred,'Residuals':(y_test-y_pred)})
```

```python
poly_MAE=metrics.mean_absolute_error(y_test,y_pred)
poly_MSE=metrics.mean_squared_error(y_test,y_pred)
poly_RMSE=np.sqrt(poly_MSE)
```
